```python
# load in the "starbucks_clean.csv" dataset
import pandas as pd

df = pd.read_csv(r"D:\Projects\MDST\Tutorial\mdst_tutorials\data\starbucks_clean.csv")
df.columns = df.columns.str.strip()
print(df.head())
```

```python
# this is nutritional info for starbucks items
# let's see if we can answer some questions

# what is the average # calories across all items?
df["calories"].mean()
```

```python
# how many different categories of beverages are there?
df["beverage_category"].nunique()
```

```python
# what is the average # calories for each beverage category?
df.groupby("beverage_category").mean()["calories"]
```

```python
# what beverage preparation includes the most sugar on average?
df.groupby("beverage_prep").mean().sort_values(by="sugars", ascending=False)
```

```python
# what is the average % daily value calcium content for each beverage?
df.groupby("beverage").mean()["calcium"]
```

```python
# It's bulking season. What drink should Naitian get so that he maximizes protein but minimizes calories?
# (you can use more than one line for this one)
# there are multiple valid approaches you can take for this (with different answers)
# I know that you want more calories for bulking season but whatever this is MDST, not Michigan Muscle Club okay

# create a new column that records protein - calories 

df["bulking_eval"] = df["protein"] - df["calories"]

df.sort_values(by="bulking_eval", ascending = False)
```

```python
# by this measure, short brewed coffee appears to be the best 
# what if we measure by the ratio between protein and calories 

df["calorie ratio"] = df["calories"]/df["protein"]
df.sort_values(by="calorie ratio")
```

```python
# this measure points to grande brewed coffee
```

```python
# plot the distribution of the number of calories in drinks using matplotlib
import matplotlib.pyplot as plt
plt.hist(df["calories"], bins=10, edgecolor="black")
plt.xlabel("Calories")
plt.ylabel("Frequency")
```

```python
# plot the same distribution using seaborn. what about a density plot instead?
import seaborn as sns 
sns.displot(data=df, x="calories", kind="kde")
```

```python
# plot calories against total fat using seaborn
sns.displot(data=df, x="total fat", y="calories")
```

```python
# plot the distribution of calories in Short, Tall, Grande, and Venti drinks
# (you can use multiple lines for this)
# twist: you should also include the Nonfat Milk drinks that also have an associated size.

# you can decide how you want to visualize this. Colors? Small multiples? Density vs histogram?
# the starbucks is your oyster.

df_short = pd.concat([df[df["beverage_prep"]=="Short"], df[df["beverage_prep"]=="Short Nonfat Milk"]])
df_tall = pd.concat([df[df["beverage_prep"]=="Tall"], df[df["beverage_prep"]=="Tall Nonfat Milk"]])
df_grande = pd.concat([df[df["beverage_prep"]=="Grande"], df[df["beverage_prep"]=="Grande Nonfat Milk"]])
df_venti = pd.concat([df[df["beverage_prep"]=="Venti"], df[df["beverage_prep"]=="Venti Nonfat Milk"]])

df_by_size = pd.concat([df_short, df_tall, df_grande, df_venti])
df_by_size["size"] = df_by_size["beverage_prep"].apply(lambda x: x[:x.find("N")-1] if "Milk" in x else x)

sns.displot(data=df_by_size, x="calories", hue="size", kind="kde")
```
